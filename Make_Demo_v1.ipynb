````python
import re

def custom_split(text):
    text_list = []
    current_char_num = 0
    text_buffer = ""
    MAX_CHAR = 400
    for x in text.split("."):
        x = x.strip()
        x = x.replace('"', "")
        x = x.replace("'", "")
        x += "."
        if current_char_num + len(x) > MAX_CHAR:
            text_list.append(text_buffer)
            text_buffer = x
            current_char_num = len(x)
        else:
            current_char_num += len(x)
            text_buffer += x
    text_list.append(text_buffer)
    return text_list

def add_space_after_dot(text):
    # Use regular expression to find dots followed by non-space characters
    # and replace them with dot followed by space
    modified_text = re.sub(r'\.(?!\s|$)', '. ', text)
    # Replace multiple spaces with a single space
    modified_text = re.sub(r'\s+', ' ', modified_text)
    return modified_text

def preproces_abbreviations(text):
    return re.sub(r'\b([A-Z]+)\b', lambda match: '-'.join(match.group(1)), text)

tts_text = custom_split(eng_text)
tts_text = [preproces_abbreviations(add_space_after_dot(text)).strip() for text in tts_text]
len(tts_text), tts_text[2]
````
